## ***2. Understanding Your Variables***

````python
## Titles Dataset (titles.csv)
| Column Name            | Description                                                                 |
|------------------------|-----------------------------------------------------------------------------|
| id                     | Unique identifier for each title                                            |
| title                  | Name of the movie or TV show                                                |
| type                   | Content type – Movie or TV Show                                             |
| description            | Short summary or synopsis of the title                                      |
| release_year           | Year the title was released                                                 |
| age_certification      | Age rating (e.g., PG, R, U/A)                                                |
| runtime                | Duration of the title in minutes                                            |
| genres                 | Genre(s) associated with the title (e.g., Drama, Comedy)                    |
| production_countries   | Country or countries that produced the title                                |
| seasons                | Number of seasons (only applicable to TV shows)                             |
| imdb_id                | IMDb identifier for linking or referencing externally                       |
| imdb_score             | IMDb rating of the title (scale of 1–10)                                     |
| imdb_votes             | Number of votes received on IMDb                                            |
| tmdb_score             | TMDB rating of the title                                                    |
| tmdb_popularity        | Popularity score from TMDB       
````

````python
## Credits Dataset (credits.csv)
| Column Name | Description                                      |
|-------------|--------------------------------------------------|
| id          | Foreign key referencing the Titles dataset       |
| name        | Name of the person involved (actor or director)  |
| character   | Character name (only for actors)                 |
| role        | Role of the person (e.g., ACTOR, DIRECTOR)       |
| person_id   | Unique ID of the person                          |
````

### Variables Description

The dataset contains metadata about Amazon Prime content, split across two CSV files:

- The *titles dataset* holds information about movies and TV shows, including their title, type (movie or show), release year, age rating, runtime, genres, production countries, and ratings from IMDb and TMDB. It also includes a short description and the number of seasons (for shows only).

- The *credits dataset* maps each title to its associated contributors. This includes the names of actors and directors, their role (e.g., ACTOR, DIRECTOR), and the characters they played (if applicable). It also contains a unique `person_id` for each contributor.

Together, these datasets allow us to explore Amazon Prime’s content catalog from multiple angles — content type, genre distribution, popularity, regional trends, and key contributors.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Drop duplicate rows from both datasets
titles_df.drop_duplicates(inplace=True)
credits_df.drop_duplicates(inplace=True)
print("Titles duplicates after cleaning:", titles_df.duplicated().sum())
print("Credits duplicates after cleaning:", credits_df.duplicated().sum())
# Fill missing values in relevant columns
titles_df['age_certification'] = titles_df['age_certification'].fillna('Not Rated')
titles_df['description'] = titles_df['description'].fillna('No description available')
credits_df['character'] = credits_df['character'].fillna('Unknown')
# Remove brackets and quotes from genres and production_countries columns
titles_df['genres'] = titles_df['genres'].str.replace(r"[\[\]']", '', regex=True)
titles_df['production_countries'] = titles_df['production_countries'].str.replace(r"[\[\]']", '', regex=True)
# Merge datasets on 'id' for full content + credits info
merged_df = pd.merge(credits_df, titles_df, on='id', how='inner')
print("Merged dataset shape:", merged_df.shape)
merged_df.head()
````

### What all manipulations have you done and insights you found?

- Removed 3 duplicate rows from the Titles dataset and 56 from the Credits dataset. After cleaning, both datasets show zero duplicates.
- Handled missing values by filling:
  - 'age_certification' with 'Not Rated'
  - 'description' with 'No description available'
  - 'character' in the Credits dataset with 'Unknown'
- Cleaned formatting for list-type columns like 'genres' and 'production_countries' by removing brackets and quotes.
- Successfully merged the Titles and Credits datasets using the common 'id' field, resulting in a final merged dataset with 124,179 rows and 19 columns.
- The merged dataset is now ready for univariate, bivariate, and multivariate visualizations.

````python
#Top 5 Genre Trends Over the Years on Amazon Prime
genre_year_df = (
    titles_df[['release_year', 'genres']]
    .dropna()
    .assign(genres=lambda df: df['genres'].str.replace(r"[\[\]']", "", regex=True))
    .assign(genres=lambda df: df['genres'].str.split(','))
    .explode('genres')
    .dropna()
    .assign(genres=lambda df: df['genres'].str.strip())
)
top_genres = genre_year_df['genres'].value_counts().nlargest(5).index
filtered_df = genre_year_df[genre_year_df['genres'].isin(top_genres)]
genre_trend = (
    filtered_df.groupby(['release_year', 'genres']).size()
    .unstack(fill_value=0)
)
genre_trend.plot(kind='area', stacked=True, figsize=(12, 6), colormap='Set2')
plt.title('Top 5 Genre Trends Over the Years on Amazon Prime')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.show()
````
